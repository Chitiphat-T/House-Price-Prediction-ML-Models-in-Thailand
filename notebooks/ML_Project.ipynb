**Data Cleaning**

```python
useful_columns = [
    'price',
    'living_space',
    'bedroom_number',
    'bathroom_number',
    'room_number',
    'parking_number',
    'property_type',
    'city',
    'state',
    'built_year',
    'furnished',
    'new_home'
]

df = df[useful_columns].copy()
df = df.drop_duplicates()
df.head()
```

```python
df.isnull().mean().sort_values(ascending = False)
```

```python
df = df.drop(columns = ['parking_number', 'new_home', 'room_number'])
df = df[df['property_type'] != 'Land']
df = df.dropna(subset=['price'])
df.isnull().mean().sort_values(ascending = False)
```

```python
df.isnull().mean().sort_values(ascending = False)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df['bathroom_number'] = pd.to_numeric(df['bathroom_number'], errors='coerce')
df['bedroom_number']  = pd.to_numeric(df['bedroom_number'], errors='coerce')
df['living_space']    = pd.to_numeric(df['living_space'], errors='coerce')
df['built_year']      = pd.to_numeric(df['built_year'], errors='coerce')
```

```python
df['bathroom_number'] = df['bathroom_number'].fillna(df['bathroom_number'].median())
df['bedroom_number']  = df['bedroom_number'].fillna(df['bedroom_number'].median())
df['living_space']    = df['living_space'].fillna(df['living_space'].median())
df = df.dropna(subset=['city', 'state', 'property_type'])

df['built_year']  = df['built_year'].fillna(df['built_year'].median())
df['furnished']   = df['furnished'].fillna('Unknown')
df = df.drop_duplicates()
```

**Outlier Detection**

```python
numerical_cols = ['price', 'living_space', 'bedroom_number', 'bathroom_number', 'built_year']
display(df[numerical_cols].describe())
```

```python
df = df[df['price'] > 500_000]
df = df[df['price'] < 100_000_000]

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 3.0 * IQR
upper_bound = Q3 + 3.0 * IQR

df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

print(f"Max price after cleaning: {df['price'].max():,.2f}")
```

```python
df = df[df['living_space'].between(10, 10000)]

df = df[df['bedroom_number'].between(1, 15)]
df = df[df['bathroom_number'].between(1, 15)]

print(f"Rows remaining after cleanup: {len(df)}")
```

**Feature Engineering**

```python
df['property_age'] = 2026 - df['built_year']

df['living_space_ratio'] = df['living_space'] / (df['bedroom_number'] + df['bathroom_number'])

df['total_rooms'] = df['bedroom_number'] + df['bathroom_number']

city_means = df.groupby('city')['price'].transform('mean')
df['city_price_index'] = city_means

type_means = df.groupby('property_type')['price'].transform('mean')
df['type_space_interaction'] = type_means * df['living_space']
```

```python
print("--- All Unique States (Provinces) ---")
states = df['state'].unique()
print(sorted(states))

print("\n--- Top 20 Cities (Districts) ---")
print(df['city'].value_counts().head(20))

print("\n--- Breakdown of Cities in the largest State ---")
top_state = df['state'].value_counts().index[0]
print(f"Top State: {top_state}")
print(df[df['state'] == top_state]['city'].value_counts().head(10))
```

```python
def get_region(state):
    central = ['Bangkok', 'Nonthaburi', 'Pathum Thani', 'Samut Prakan', 'Samut Sakhon', 'Nakhon Pathom']
    east = ['Chon Buri', 'Rayong', 'Chachoengsao', 'Pattaya']
    north = ['Chiang Mai', 'Chiang Rai', 'Lampang', 'Phuket', 'Krabi'] # Phuket/Krabi are South but often grouped for high-tourism
    if state in central: return 'BKK_Metropolitan'
    if state in east: return 'EEC_Zone'
    if state in ['Phuket', 'Krabi', 'Surat Thani', 'Phangnga']: return 'Southern_Tourist'
    return 'Upcountry'

df['region'] = df['state'].apply(get_region)

df['city'] = df['city'].replace({'Pattaya': 'Bang Lamung'})

bangkok_metro = [
    'Bangkok', 'Krung Thep Maha Nakhon', 'Nonthaburi',
    'Samut Prakan', 'Pathum Thani', 'Samut Sakhon', 'Nakhon Pathom'
]

df['is_bangkok'] = df['state'].apply(lambda x: 1 if x in bangkok_metro else 0)
```

```python
city_sqm_price = df.groupby('city').apply(lambda x: x['price'].sum() / x['living_space'].sum()).to_dict()
df['city_sqm_baseline'] = df['city'].map(city_sqm_price)

city_avg_size = df.groupby('city')['living_space'].transform('mean')
df['size_vs_city_avg'] = df['living_space'] / city_avg_size

df['space_per_bed'] = df['living_space'] / df['bedroom_number'].replace(0, 1)

df['is_micro_unit'] = (df['living_space'] < 35).astype(int)
```

**Data Preprocessing**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import category_encoders as ce
from category_encoders import TargetEncoder
from sklearn.pipeline import Pipeline
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
num = [
    'living_space', 'bedroom_number', 'bathroom_number', 'built_year',
    'property_age', 'living_space_ratio', 'total_rooms',
    'city_price_index', 'type_space_interaction',
    'city_sqm_baseline', 'size_vs_city_avg', 'space_per_bed',
]

cat_subset = [
    'property_type',
    'state',
    'furnished',
    'region',
    'is_bangkok'
]
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_subset),
        ('city_enc', TargetEncoder(), ['city'])
    ]
)
```

```python
pipeline_lr = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', LinearRegression())
])

pipeline_dt = Pipeline(steps=[
    ('preprocessor', preprocess),
    ('regressor', DecisionTreeRegressor(max_depth = 15, min_samples_leaf = 50, random_state = 42
    ))
])

pipeline_xgb = Pipeline(steps=[
    ('preprocessor', preprocess),
    ('regressor', XGBRegressor(
        objective='count:poisson',
        n_estimators=2500,
        learning_rate=0.02,
        max_depth=10,
        random_state=42
    ))
])
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size = 0.2,
    random_state = 42,
)
```

**Linear Regression**

```python
pipeline_lr.fit(x_train, y_train)
pred_lr = pipeline_lr.predict(x_test)

r2_lr = r2_score(y_test, pred_lr)
mae_lr = mean_absolute_error(y_test, pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, pred_lr))

print("--- Linear Regression ---")
print(f"RÂ²:   {r2_lr:.4f}")
print(f"MAE:  {mae_lr:,.2f}")
print(f"RMSE: {rmse_lr:,.2f}")
```

**Decision Tree**

```python
pipeline_dt.fit(x_train, y_train)
pred_rf = pipeline_dt.predict(x_test)

r2_dt = r2_score(y_test, pred_rf)
mae_dt = mean_absolute_error(y_test, pred_rf)
rmse_dt = np.sqrt(mean_squared_error(y_test, pred_rf))

print("--- Decision Tree ---")
print(f"RÂ²:   {r2_dt:.4f}")
print(f"MAE:  {mae_dt:,.2f}")
print(f"RMSE: {rmse_dt:,.2f}")
```

**XGBoost**

```python
pipeline_xgb.fit(x_train, y_train)
pred_xgb = pipeline_xgb.predict(x_test)

r2_xgb = r2_score(y_test, pred_xgb)
mae_xgb = mean_absolute_error(y_test, pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, pred_xgb))

print("--- XGBoost ---")
print(f"RÂ²:   {r2_xgb:.4f}")
print(f"MAE:  {mae_xgb:,.2f}")
print(f"RMSE: {rmse_xgb:,.2f}")
```

**Results**

```python
results_df = pd.DataFrame([
    ['Linear Regression', rmse_lr, mae_lr, r2_lr],
    ['Decision Tree', rmse_dt, mae_dt, r2_dt],
    ['XGBoost', rmse_xgb, mae_xgb, r2_xgb],
], columns = ['Model', 'RMSE', 'MAE', 'RÂ²']
)

results_df = results_df.sort_values(by='RÂ²', ascending=False).reset_index(drop=True)

results_df.round(2)
```
